```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('ticks')

data=np.loadtxt("data.txt",skiprows=1,delimiter=',')
```

```python
for i in range(1,data.shape[0]+1):
    fig=plt.figure(figsize=(10,5))
    with sns.axes_style('darkgrid',{'font.family':'DejaVu Sans'}):
    
        sns.despine()
        ax1=fig.add_subplot(121)
        plt.plot(data[0:i,0],data[0:i,1],"o")
        plt.axvline(x=0,alpha=.1,color='black')
        plt.axhline(y=0,alpha=.1,color='black')
        plt.xlim(-125,125)
        plt.ylim(-125,125)
        fig.axes[0].spines['left'].set_position('center')
        fig.axes[0].spines['bottom'].set_position('center')
        ax1.set_xticks([-100,-50,0,50,100])
        ax1.set_yticks([-100,-50,0,50,100])
        ax1.set_xticklabels([-100,-50,'',50,100])
        ax1.set_yticklabels([-100,-50,'',50,100])
        plt.title("Pencil Landings")
    
    
    with sns.axes_style('darkgrid',{'font.family':'DejaVu Sans'}):  
        ax2=fig.add_subplot(122) 
        sns.despine(left=True,bottom=True)
        r=np.sqrt((data[0:i,0])**2+(data[0:i,1])**2)
        plt.hist(r,bins=10,range=(0,100),rwidth=.9)
        plt.ylim(0,20)
        ax2.set_yticklabels([])
        ax2.set_xlabel("r (cm)")
        plt.title("Radial Distribution")
        
    plt.savefig('pencil_figs/pencil_data_prac{:02d}.png'.format(i))
```

```python
anim_file = 'pencil_prac.gif'
import glob
import imageio

with imageio.get_writer(anim_file, mode='I') as writer:
    filenames = glob.glob('pencil_figs/pencil*.png')
    filenames = sorted(filenames)
    last = -1
    for i,filename in enumerate(filenames):     
        image = imageio.imread(filename)
        writer.append_data(image)
```
